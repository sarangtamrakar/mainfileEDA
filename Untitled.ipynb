```python
class impute:
    
    """ THIS CLASS SPECIALLY DESIGNED FOR IMPUTE THE MISSING VALUES BASED ON DIFFERENT STRETEGIES..."""
    
    def __init__(self):
        pass
    
    
    
    def impute_by_mode(self,df,var):
        
        """ THIS METHOD IS USED FOR CATEGORYCAL COLS & WHICH IS HAVING MISSING PCT IS LESS THAN 20% """
        
        mode = df[var].mode()[0]
        df[var] = df[var].fillna(mode)
    
    
    
    
    def impute_by_median(self,df,var):
        
        """THIS METHOD IS USED FOR CONTINUOUS COLS & WHICH IS HAVING MISSING_PCT IS LESS THAN 20% """
        
        median = df[var].median()
        df[var] = df[var].fillna(median)
        
    
    
    def impute_by_capture_nan_in_continuous_cols(self,df,var):
        
        """ THIS METHOD IS USED FOR CONTINUOUS COL & WHICH IS HAVING MISSING_PCT IS >= 20% """
        
        median_cap = df[var].median()
        df[var+"_nan"] = np.where(df[var].isnull(),1,0)
        df[var] = df[var].fillna(median_cap)
        
        
        
        
        
        
        
obj = impute()

for col in cat_cols:
    obj.impute_by_mode(df,col)
    
for col in continuous_col:
    obj.impute_by_median(df,col)

obj.impute_by_capture_nan_in_continuous_cols(df,"T3")
```

##### 1 where the categories are 2 so we will use get dummy function of pandas..

```python
from sklearn.preprocessing import LabelEncoder
import pickle

# get col_list with 2 unique values
two_cat_list = []
for col in cat_cols:
    if df[col].nunique() == 2:
        two_cat_list.append(col)
print(two_cat_list)
```

```python
class encoding:
    
    """ THIS CLASS SPECIALLY DESIGNED FOR ENCODING THE CATEGORICAL COLUMNS..."""
    
    def __init__(self):
        pass
    
    def get_dummy_for_two_cat(self,columns_list,xdata):
        self.new_data = pd.get_dummies(data=xdata,prefix_sep='_',columns=columns_list,drop_first=True)
        return self.new_data
    
    def get_dummy_for_more_than_two_cat(self,columns_list,xdata):
        self.df = pd.get_dummies(data=xdata,prefix_sep='_',columns=columns_list,drop_first=True)
        return self.df
        
    def encode_class_label_ytrain(self,ytrain):
        self.le = LabelEncoder().fit(ytrain)
        self.ytrain = self.le.transform(ytrain)
        
        with open("class_encoder.sav","wb") as f:
            pickle.dump(self.le,f)
            
        return self.ytrain
    
    def encode_class_label_ytest(self,ytest):
        enco = pickle.load(open("class_encoder.sav","rb"))
        self.ytest=enco.transform(ytest)
        return self.ytest
    
```

```python
obj = encoding()

# encoding cat feature with two categories..
xtrain=obj.get_dummy_for_two_cat(columns_list=two_cat_list,xdata=xtrain)
xtest = obj.get_dummy_for_two_cat(columns_list=two_cat_list,xdata=xtest)

# encoding cat nominal feature with more than two......
xtrain = obj.get_dummy_for_more_than_two_cat(columns_list=["referral_source"],xdata=xtrain)
xtest = obj.get_dummy_for_more_than_two_cat(columns_list=["referral_source"],xdata=xtest)

# encoding the class label with LabelEncoder...
ytrain = obj.encode_class_label_ytrain(ytrain=ytrain)
ytest = obj.encode_class_label_ytest(ytest=ytest)
```

#### IMP NOTE.. when we have imbalanced data set..


F1 MACRO.. (it calculates the fi score for  each class then avg it then give the f1 metric..
            it is not influenced by Imbalanced dataset)

```python
class handle_class_imbalance:
    
    def __init__(self):
        pass
    
    
    def performance_xg(self,xtrain,ytrain,xtest,ytest):
        xg = XGBClassifier(random_state=13)
        xg.fit(xtrain,ytrain)
        ypred = xg.predict(xtest)
        ypred_proba = xg.predict_proba(xtest)
    
        roc = roc_auc_score(ytest,ypred_proba,multi_class='ovr')
        print("\n","ROC",roc)
    
    def performance_rf(self,xtrain,ytrain,xtest,ytest):
        xg = RandomForestClassifier(random_state=11,class_weight="balanced") #we are apply the class_weight technique..
        xg.fit(xtrain,ytrain)
        ypred = xg.predict(xtest)
        ypred_proba = xg.predict_proba(xtest)
    
        roc = roc_auc_score(ytest,ypred_proba,multi_class='ovr')
        print("\n","ROC",roc)
    
    
    def over_sampling(self,xtrain,ytrain,xtest,ytest):
        os = RandomOverSampler()
        xtrain_os,ytrain_os = os.fit_resample(xtrain,ytrain)
        result_xg = self.performance_xg(xtrain_os,ytrain_os,xtest,ytest)
        result_rf = self.performance_rf(xtrain_os,ytrain_os,xtest,ytest)
        return result_xg,result_rf
    
    def over_sampling_final_output(self,xtrain,ytrain):
        os = RandomOverSampler()
        xtrain_final,ytrain_final = os.fit_resample(xtrain,ytrain)
        return xtrain_final,ytrain_final
```

```python
# performance without sampling...

obj = handle_class_imbalance()
obj.performance_xg(xtrain,ytrain,xtest,ytest)
obj.performance_rf(xtrain,ytrain,xtest,ytest)
```

```python
# performance with over sampling... & random forest with class weight technique..
obj.over_sampling(xtrain,ytrain,xtest,ytest)
```

```python
from sklearn.feature_selection import chi2,f_classif,mutual_info_classif,RFE,RFECV,SelectFromModel,SelectKBest
```

here we already checked  the score thats why selected k =12 
& in selectfrom model set parameter max_feature

```python
class feature_selection:
    
    """this class specially designed for feature selection """
    def __init__(self):
        pass
    
    def performance(self,xtrain,ytrain,xtest,ytest):
        xg_clf = XGBClassifier()
        xg_base_line = xg_clf.fit(xtrain,ytrain)
        ypred = xg_base_line.predict(xtest)
        ypred_proba = xg_base_line.predict_proba(xtest)
        roc = roc_auc_score(ytest,ypred_proba,multi_class="ovr")
        fscore = f1_score(ytest,ypred,average='macro')
        
        print("fscore:",fscore)
        print("roc:",roc)
    
    def chi2(self,xtrain,ytrain,xtest,ytest):
        fs = SelectKBest(score_func=chi2,k=12)
        fs.fit(xtrain,ytrain)
        xtrain_chi = fs.transform(xtrain)
        xtest_chi = fs.transform(xtest)
        result_chi = self.performance(xtrain_chi,ytrain,xtest_chi,ytest)
        return result_chi
    
    def ANOVA(self,xtrain,ytrain,xtest,ytest):
        an = SelectKBest(score_func=f_classif,k=12)
        an.fit(xtrain,ytrain)
        xtrain_ano = an.transform(xtrain)
        xtest_ano = an.transform(xtest)
        result_an = self.performance(xtrain_ano,ytrain,xtest_ano,ytest)
        return result_an
    
    
    def mutual_info(self,xtrain,ytrain,xtest,ytest):
        mu = SelectKBest(score_func=mutual_info_classif,k=12)
        mu.fit(xtrain,ytrain)
        xtrain_mu = mu.transform(xtrain)
        xtest_mu = mu.transform(xtest)
        result_mu = self.performance(xtrain_mu,ytrain,xtest_mu,ytest)
        return result_mu
        
        
    def rfe(self,xtrain,ytrain,xtest,ytest):
        rfe = RFE(estimator=XGBClassifier(),n_features_to_select=12, step=1,)
        rfe.fit(xtrain,ytrain)
        xtrain_rfe = rfe.transform(xtrain)
        xtest_rfe = rfe.transform(xtest)
        result_rfe = self.performance(xtrain_rfe,ytrain,xtest_rfe,ytest)
        return result_rfe
    
    def selectfrommodel_featureimp(self,xtrain,ytrain,xtest,ytest,max_feature):
        xg = XGBClassifier()
        sfm = SelectFromModel(estimator=xg,max_features=max_feature)
        sfm_fit = sfm.fit(xtrain,ytrain)
        xtrain_sfm = sfm_fit.transform(xtrain)
        xtest_sfm = sfm_fit.transform(xtest)
        result_sfm = self.performance(xtrain_sfm,ytrain,xtest_sfm,ytest)
        return result_sfm
    
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC,LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold,cross_val_score
from sklearn.metrics import SCORERS
```

```python
class model_selection:
    def __init__(self,xtrainchi,ytrain,xtestchi,ytest):
        self.xtrain = xtrainchi
        self.ytrain = ytrain
        self.xtest = xtestchi
        self.ytest = ytest
        
    
    def Logisticregression(self):
        logit = LogisticRegression().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("logistict:",f_score)
        
    def adaboost(self):
        logit = AdaBoostClassifier().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("adaboost:",f_score)
    
    def gradientboost(self):
        logit = GradientBoostingClassifier().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("gradientboost:",f_score)
        
    def randomforest(self):
        logit = RandomForestClassifier().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("randomforest:",f_score)
        
    def xgboost(self):
        logit = XGBClassifier().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("xgboost:",f_score)
        
    def knn(self):
        logit = KNeighborsClassifier().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("knn:",f_score)
        
    def svc(self):
        logit = SVC().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("kernelsvc:",f_score)
    
    def linearsvc(self):
        logit = LinearSVC().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("linearsvc:",f_score)
        
    def decisiontree(self):
        logit = DecisionTreeClassifier().fit(self.xtrain,self.ytrain)
        ypred = logit.predict(self.xtest)
        f_score = f1_score(self.ytest,ypred,average='macro')
        print("decisiontree:",f_score)
        
    def all_classifier_score(self):
        self.Logisticregression()
        self.adaboost()
        self.decisiontree()
        self.gradientboost()
        self.knn()
        self.linearsvc()
        self.randomforest()
        self.svc()
        self.xgboost()
```
